### Part 1: TD Prediction: State Values

In this section, you will write your own implementation of TD prediction (for estimating the state-value function).

We will begin by investigating a policy where the agent moves:
- `RIGHT` in states `0` through `10`, inclusive,  
- `DOWN` in states `11`, `23`, and `35`, and
- `UP` in states `12` through `22`, inclusive, states `24` through `34`, inclusive, and state `36`.

The policy is specified and printed below.  Note that states where the agent does not choose an action have been marked with `-1`.

````python
policy = np.hstack([1*np.ones(11), 2, 0, np.zeros(10), 2, 0, np.zeros(10), 2, 0, -1*np.ones(11)])
print("\nPolicy (UP = 0, RIGHT = 1, DOWN = 2, LEFT = 3, N/A = -1):")
print(policy.reshape(4,12))
````

Run the next cell to visualize the state-value function that corresponds to this policy.

The above figure is what you will try to approximate through the TD prediction algorithm.

Your algorithm for TD prediction has five arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `policy`: This is a 1D numpy array with `policy.shape` equal to the number of states (`env.nS`).  `policy[s]` returns the action that the agent chooses when in state `s`.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `V`: This is a dictionary where `V[s]` is the estimated value of state `s`.

Please complete the function in the code cell below.

````python
from collections import defaultdict, deque
import sys

def td_prediction(env, num_episodes, policy, alpha, gamma=1.0):
    # initialize empty dictionaries of floats
    V = defaultdict(float)
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        # initialize episode instance (State = 36)
        state = env.reset()
        # begin time step loop
        while True:
            # choose action based on policy
            action = policy[state]
            # take action and record next state, reward and if next state is terminal
            next_state, reward, terminal_s, other = env.step(action)
            # update state values of corresponding to policy
            V[state] = V[state] + alpha * (reward + (gamma * V[next_state]) - V[state])
            # update state after taking action
            state = next_state
            # end time step loop if state is a terminal state
            if terminal_s:
                break
    return V 
````

### Part 2: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
# create an action-prob function using epsilon-greedy policy
def epsilon_action(env, Q_state, i_episode, eps=None):
    # define epsilon to linearly decrease w/ episode index
    epsilon = 1.0 / i_episode
    # if user decides to explicitly set epsilon
    if eps is not None:
        epsilon = eps
    # initialize action policy to exploration probability
    policy_s = np.ones(env.nA) * epsilon / env.nA
    # select index of highest action value given current state
    best_action = np.argmax(Q_state)
    # replace exploration probability for best action with exploitation prob.
    policy_s[best_action] = 1 - epsilon + epsilon / env.nA
    # return policy action probabilities
    return policy_s
````

````python
# import necessary plotting libraries
import matplotlib.pyplot as plt
%matplotlib inline

def sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    plot_every = 100
    # latest_scores will be a running list of latest plot_every scores 
    latest_scores = deque(maxlen=plot_every)
    # avg_scores will be the average of latest_scores after every plot_every latest_scores
    every_100_score_avg = []
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()  
        # initialize score
        score = 0
        # initialize an episode (State 0)
        state = env.reset()
        # get epsilon-greedy action probabilities for State 0
        policy_s = epsilon_action(env, state, i_episode)
        # select action given probabilities
        action = np.random.choice(np.arange(env.nA), p=policy_s)
        # limit time steps for a given episode
        for t_step in np.arange(300):
            # take action and record next state, reward and if terminal state
            next_state, reward, terminal_s, other = env.step(action)
            # add time step reward to score
            score += reward
            # if action does not lead to terminal state, continue to next time step
            if not terminal_s:
                # get next action probabilities
                policy_s = epsilon_action(env, Q[next_state], i_episode)
                # select random next action given action probabilities
                next_action = np.random.choice(np.arange(env.nA), p=policy_s)
                # current action_value
                current_q = Q[state][action]
                # next action value
                next_q = Q[next_state][next_action]
                # update current action value 
                Q[state][action] = current_q + alpha * (reward + gamma * next_q - current_q)
                # update state
                state = next_state
                # update action
                action = next_action
            # if action leads to terminal state, append episode score to latest_scores
            if terminal_s:
                # update action values at current state-action
                Q[state][action] += alpha * (reward - Q[state][action])
                # add score to latest scores
                latest_scores.append(score)
                # stop time steps
                break
        # add latest scores average to avg_scores
        # if episode number is divisible by plot_every
        if i_episode % plot_every == 0:
            every_100_score_avg.append(np.mean(latest_scores))
    # plot avg_scores
    # define x_axis
    x_axis = np.linspace(0, num_episodes, len(every_100_score_avg), endpoint=False)
    # plot scores
    plt.plot(x_axis, np.asarray(every_100_score_avg))
    plt.xlabel('Episode Number')
    plt.ylabel('Average Reward')
    plt.show()
    
    # display best average reward
    print('Best Average Reward:\t', np.max(every_100_score_avg))
    return Q
````

### Part 3: TD Control: Q-learning

In this section, you will write your own implementation of the Q-learning control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.

````python
# import necessary plotting libraries
import matplotlib.pyplot as plt
%matplotlib inline

def q_learning(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    plot_every = 100
    # latest_scores will be a running list of latest plot_every scores 
    latest_scores = deque(maxlen=plot_every)
    # avg_scores will be the average of latest_scores after every plot_every latest_scores
    every_100_score_avg = []
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()  
        # initialize score
        score = 0
        # initialize an episode (State 0)
        state = env.reset()
        # begin episode
        while True:
            # get epsilon-greedy action probabilities for State 0
            policy_s = epsilon_action(env, Q[state], i_episode)
            # select action given probabilities
            action = np.random.choice(np.arange(env.nA), p=policy_s)
            # take action and record next state, reward and if terminal state
            next_state, reward, terminal_s, other = env.step(action)
            # add time step reward to score
            score += reward
            # current action_value
            current_q = Q[state][action]
            # next action value
            max_next_q = np.max(Q[next_state])
            # update current action value 
            Q[state][action] = current_q + alpha * (reward + gamma * max_next_q - current_q)                
            # update state
            state = next_state
            # if action leads to terminal state, append episode score to latest_scores
            if terminal_s:
                # add score to latest scores
                latest_scores.append(score)
                # stop time steps
                break
        # add latest scores average to avg_scores
        # if episode number is divisible by plot_every
        if i_episode % plot_every == 0:
            every_100_score_avg.append(np.mean(latest_scores))
    # plot avg_scores
    # define x_axis
    x_axis = np.linspace(0, num_episodes, len(every_100_score_avg), endpoint=False)
    # plot scores
    plt.plot(x_axis, np.asarray(every_100_score_avg))
    plt.xlabel('Episode Number')
    plt.ylabel('Average Reward')
    plt.show()
    
    # display best average reward
    print('Best Average Reward:\t', np.max(every_100_score_avg))
    return Q
````

### Part 4: TD Control: Expected Sarsa

In this section, you will write your own implementation of the Expected Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
# import necessary plotting libraries
import matplotlib.pyplot as plt
%matplotlib inline

def expected_sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    plot_every = 100
    # latest_scores will be a running list of latest plot_every scores 
    latest_scores = deque(maxlen=plot_every)
    # avg_scores will be the average of latest_scores after every plot_every latest_scores
    every_100_score_avg = []
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()  
        # initialize score
        score = 0
        # initialize an episode (State 0)
        state = env.reset()
        # begin episode
        while True:
            # get epsilon-greedy action probabilities for State 0
            policy_s = epsilon_action(env, Q[state], i_episode, 0.0005)
            # select action given probabilities
            action = np.random.choice(np.arange(env.nA), p=policy_s)
            # take action and record next state, reward and if terminal state
            next_state, reward, terminal_s, other = env.step(action)
            # add time step reward to score
            score += reward
            # current action_value
            current_q = Q[state][action]
            # next state action probabilities
            policy_ns = epsilon_action(env, Q[next_state], i_episode, 0.0005)
            # set expected action value using all possible next action values and their corresponding probabilities
            q_all = np.dot(Q[next_state], policy_ns)
            # update current action value 
            Q[state][action] = current_q + alpha * (reward + gamma * q_all - current_q)                
            # update state
            state = next_state
            # if action leads to terminal state, append episode score to latest_scores
            if terminal_s:
                # add score to latest scores
                latest_scores.append(score)
                # stop time steps
                break
        # add latest scores average to avg_scores
        # if episode number is divisible by plot_every
        if i_episode % plot_every == 0:
            every_100_score_avg.append(np.mean(latest_scores))
    # plot avg_scores
    # define x_axis
    x_axis = np.linspace(0, num_episodes, len(every_100_score_avg), endpoint=False)
    # plot scores
    plt.plot(x_axis, np.asarray(every_100_score_avg))
    plt.xlabel('Episode Number')
    plt.ylabel('Average Reward')
    plt.show()
    
    # display best average reward
    print('Best Average Reward:\t', np.max(every_100_score_avg))
    return Q
````
